```python
import pandas as pd
import numpy as np
import os

file_format = ".xlsx" # .csv
file_path = "C:/박규영/축구포트폴리오/2023K리그데이터포털"
file_list = [f"{file_path}/{file}" for file in os.listdir(file_path) if file_format in file]
```

```python
# 모든 파일을 담을 빈 리스트
all_dataframes = []

# 파일을 순회하며 DataFrame으로 읽어와 리스트에 추가
for file in file_list:
    df = pd.read_excel(file)
    all_dataframes.append(df)

# 리스트에 있는 모든 DataFrame을 하나로 합침
merged_df = pd.concat(all_dataframes, ignore_index=True)
```

- 따라서 기존 데이터프레임 컬럼명 사용

```python
merged_df = pd.read_csv("C:/박규영/축구포트폴리오/2023K리그데이터포털/전처리/2023K리그데이터통합.csv", encoding = 'cp949')
```

```python
# 성공률 제거한 이유, 간단한 계산으로 만들 수 있음
# 아래 전처리 과정에서 sum을 기준으로 집계할건데 성공률은 sum하면 안됨

merged_df = merged_df.drop(columns=[col for col in merged_df.columns if '성공률' in col])
```

#### 2. 선수 포지션 처리
- 목표1: CB, MF, FW, GK -> CB, FB, MF, FW, GK로 세분화
- 목표2: 데이터프레임의 '포지션 컬럼'의 구분을 보면 같은 선수여도 'FW', '대기' 이런식으로 있음
    - 대구의 고재현 같은 경우, FW로 나올 때, MF로 나올 때가 구분되어 있음
    - 경남의 박민서는 두 명으로 같은 팀의 동명이인이 존재할 수도 있음
- 따라서 주 포지션 하나로 묶던가, 동일한 선수별 고유번호를 만드는 방법을 사용해야 함
- 고유번호를 만드는 조건생각해야 함

```python
seoul = merged_df[merged_df['구단'] == 'FC서울']

print('고유 선수 숫자:', len(seoul['선수명'].unique()))
print('FC서울 데이터프레임 내 인덱스 숫자:', len(seoul))
print(seoul['선수명'].unique())
```

```python
# 선수명, 구단, 번호를 기준으로 합침 

new_df = merged_df.groupby(['선수명','구단','번호','리그'], as_index=False).sum()
len(new_df)
```

문제는 팀명이 없어서 동명이인 처리가 빡세다..

```python
data = merged_df.drop_duplicates(subset=['선수명', '구단', '번호'], keep='first')
len(data)
```

```python
new_df[new_df['선수명']=='박민서'] # 포지션 x, 경기 기록 제대로 합산
```

```python
data[data['선수명']=='박민서'] # 포지션 ㅇ, 경기 기록 제대로 합산 x
```

new_df와 data라는 데이터프레임의 구단, 선수명, 번호, 리그를 기준으로 new_df의 경기 기록 데이터에 data의 포지션 컬럼을 추가 

```python
s = pd.merge(new_df, data[['구단', '선수명', '번호', '리그', '포지션']], on=['구단', '선수명', '번호', '리그'], how='left')
s[s['선수명']=='박민서']
```

### 3. 선수인적정보 추가

```python
info = pd.read_excel("C:/박규영/축구포트폴리오/2023K리그데이터포털/인적정보/2023_K리그_선수전체인적정보.xlsx")
```

```python
info = info.iloc[:,2:]
info['번호'] = info['LABEL-3'].str.split('.').str[0]
info['선수명'] = info['LABEL-3'].str.split('.').str[-1]
info['신장'] = info['LABEL-4'].str.split('/').str[0]
info['생년월일'] = info['LABEL-4'].str.split('/').str[2]
```

```python
info = info.iloc[:,2:]
info['번호'] = info['번호'].astype('int64')
info['선수명'] = info['선수명'].str.strip()
```

```python
# 선수명에 공백 제거를 했어야 함

info[info['선수명'] == '김정호']
```

```python
k = pd.merge(s, info, how = 'left', on = ['선수명','번호'])
```

```python
k[k['생년월일'].isnull()]
```

- 또한, 시즌 중반에 이적한 선수들의 경우 선수명+번호를 이용해 join을 하는 방식인데 번호가 일치하지않아 결측치가 생김
